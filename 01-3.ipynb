# k-최근접 이웃 알고리즘

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
# 두개를 섞어서 분류해내는것을 할려면 [[길이,무게], [길이,무게], ...] 이런식으로 데이터를 만들어야한다. (2차원 리스트)
fish_data = [[l,w] for l ,w in zip(length, weight)]
target = [1]*35 + [0]*14
# 도미는 1, 빙어는 0으로 타겟값을 만들어준다.
# 우리가 찾아내길 원하는 생선은 도미이기 때문이다.
# 그러니까 머신에게 학습을 시킨다. 
```
